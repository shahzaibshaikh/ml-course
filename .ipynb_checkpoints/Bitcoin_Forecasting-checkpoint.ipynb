```python
import torch
import numpy as np
import pandas as pd
import shutil

from darts import TimeSeries
from darts.models import NHiTSModel, TiDEModel, TCNModel, BlockRNNModel
from darts.datasets import AusBeerDataset
from darts.dataprocessing.transformers.scaler import Scaler
from pytorch_lightning.callbacks.early_stopping import EarlyStopping
from darts.metrics import mae, mse

import matplotlib.pyplot as plt

import warnings

warnings.filterwarnings("ignore")
import logging
logging.disable(logging.CRITICAL)
```

```python
import copy
from pathlib import Path

# List of columns you want to read
selected_columns = ["date", "open", "close"]

file_path = 'BTC-2021min.csv'
df = pd.read_csv(file_path, usecols=selected_columns)
df.head(-1)
```

```python
# Convert 'Timestamp [ms]' to datetime and set it as the index
df['date'] = pd.to_datetime(df['date'])
df.set_index('date', inplace=True)

df = df.resample('15min').mean()

# Preserve the 'Timestamp [ms]' column
df['date'] = df.index

df.head()
```

```python
# Finding the median value from open and close prices
df["median"] = (df["open"] + df["close"])/2

# Filter to keep only data after 2021-11
df = df[df['date'] >= '2021-11-01']
# df = df[df['date'] <= '2022-01-20']

df.head(-1)
```

```python
# Create a Darts TimeSeries object
series = TimeSeries.from_dataframe(df, "date", 'median', freq="15min")

print(len(series))
```

```python
train, temp = series.split_after(0.7)
val, test = temp.split_after(0.5)

plt.figure(figsize=(24, 6))
train.plot(label="train")
val.plot(label="val")
test.plot(label="test")
```

```python
import pytorch_lightning as pl
from pytorch_lightning.loggers import TensorBoardLogger
from pytorch_lightning.callbacks.early_stopping import EarlyStopping

scaler = Scaler()  # default uses sklearn's MinMaxScaler
train = scaler.fit_transform(train)
val = scaler.transform(val)
test = scaler.transform(test)

plt.figure(figsize=(24, 6))
train.plot()
val.plot()
test.plot()
```

```python
# Common parameters
common_model_args = {
    "input_chunk_length": 672,
    "output_chunk_length": 336,
    "optimizer_kwargs": {"lr": 1e-3},
    "pl_trainer_kwargs": {
        "gradient_clip_val": 1,
        "max_epochs": 20,
        "accelerator": "auto",
        "callbacks": [],
        "logger": None
    },
    "lr_scheduler_cls": torch.optim.lr_scheduler.ExponentialLR,
    "lr_scheduler_kwargs": {"gamma": 0.999},
    "likelihood": None,
    "save_checkpoints": True,
    "force_reset": True,
    "batch_size": 64, #32
    "random_state": 42,
}

# Define early stopping arguments
early_stopping_args = {
    "monitor": "val_loss",
    "patience": 3,
    "mode": "min",
    "strict": True,
    "verbose": True,
}

# Update your pl_trainer_kwargs
pl_trainer_kwargs = common_model_args["pl_trainer_kwargs"]


# Model-specific parameters
model_nhits_params = {
    "model_name": "nhits_bitcoin",
    "num_stacks": 6,
    "num_blocks": 2,
}

model_tide_params = {
    "model_name": "tide_bitcoin",
    "use_reversible_instance_norm": False,
    "dropout": 0.3
}

model_tcn_params = {
    "model_name": "tcn_bitcoin",
    "num_filters": 6,
    "kernel_size": 6
}

model_lstm_params = {
    "model_name": "lstm_bitcoin",
    "model": "LSTM",
    "n_rnn_layers": 4
}

# Define each model by merging common and model-specific parameters
model_nhits = NHiTSModel(**common_model_args, **model_nhits_params)
model_tide = TiDEModel(**common_model_args, **model_tide_params)
model_tcn = TCNModel(**common_model_args, **model_tcn_params)
model_lstm = BlockRNNModel(**common_model_args, **model_lstm_params)


models = {
    "NHiTS": model_nhits,
    "TiDE": model_tide,
    "TCN": model_tcn,
    # "LSTM": model_lstm
}
```

```python
# Saving the model to disk
import pickle

# Set the log directory where logs will be saved
log_dir = "logs"

# Create a TensorBoardLogger
logger = TensorBoardLogger(log_dir, name="bitcoin_all_models")

# Update your pl_trainer_kwargs
pl_trainer_kwargs["logger"] = logger

# Train the models and load the model from its best state/checkpoint
for name, model in models.items():
    # early stopping needs to get reset for each model
    pl_trainer_kwargs["callbacks"] = [
        EarlyStopping(
            **early_stopping_args,
        )
    ]

    # Use the logger in the Trainer (remove 'logger' from Trainer arguments)
    pl_trainer = pl.Trainer(**pl_trainer_kwargs)

    model.fit(
        series=train,
        val_series=val,
        verbose=True,
    )
    
    model.save(f"models/{name}_bitcoin_forecasting_min.pt")
```

```python
# Loading the saved model
model_nhits = NHiTSModel.load('models/NHiTS_bitcoin_forecasting_min.pt')
model_tide = TiDEModel.load('models/TiDE_bitcoin_forecasting_min.pt')
model_tcn = TCNModel.load('models/TCN_bitcoin_forecasting_min.pt')
model_lstm = BlockRNNModel.load('models/LSTM_bitcoin_forecasting_min.pt')

# Predicting for the next 2 weeks
pred_steps = 286
pred_input = test[:-pred_steps]

# Performing the prediction
pred_series_nhits = model_nhits.predict(n=pred_steps, series=pred_input)
pred_series_tide = model_tide.predict(n=pred_steps, series=pred_input)
pred_series_tcn = model_tcn.predict(n=pred_steps, series=pred_input)

# Calculating error rates
mae_result_nhits = mae(test, pred_series_nhits)
mae_result_tide = mae(test, pred_series_tide)
mae_result_tcn = mae(test, pred_series_tcn)

mse_result_nhits = mse(test, pred_series_nhits)
mse_result_tide = mse(test, pred_series_tide)
mse_result_tcn = mse(test, pred_series_tcn)


# Inverse transforming signals to bring them to original scale
pred_series_nhits = scaler.inverse_transform(pred_series_nhits)
pred_series_tide = scaler.inverse_transform(pred_series_tide)
pred_series_tcn = scaler.inverse_transform(pred_series_tcn)
pred_input = scaler.inverse_transform(pred_input)
test = scaler.inverse_transform(test)

# Converting to dataframe
pred_input = pred_input.pd_dataframe()
test_series = test.pd_dataframe()
pred_series_nhits = pred_series_nhits.pd_dataframe()
pred_series_tide = pred_series_tide.pd_dataframe()
pred_series_tcn = pred_series_tcn.pd_dataframe()

# Restoring saved Timestamp column from index values
pred_input["date"] = pred_input.index
test_series["date"] = test_series.index
pred_series_nhits["date"] = pred_series_nhits.index
pred_series_tide["date"] = pred_series_tide.index
pred_series_tcn["date"] = pred_series_tcn.index
pred_series_lstm["date"] = pred_series_lstm.index

# Set the specific start date for plotting
start_date = pd.Timestamp("2022-02-17")
end_date = pd.Timestamp("2022-03-02")

# Create subplots
fig, axs = plt.subplots(3, 1, figsize=(28, 18))
axs = axs.flatten()

# Plotting NHITS
axs[0].plot(pred_input['date'], pred_input['median'], label='Input', linewidth=3)
axs[0].plot(test_series['date'][-pred_steps:], test_series['median'][-pred_steps:], label='Ground Truth', linewidth=3)
axs[0].plot(pred_series_nhits['date'], pred_series_nhits['median'], label='NHITS', linewidth=3)
axs[0].set_title('NHITS', fontweight='bold')
axs[0].set_xlabel('Timestamp (H)')
axs[0].set_ylabel('Price (USD)')
axs[0].legend()
axs[0].grid(False, axis='x')
axs[0].set_xlim(start_date, end_date)

# Plotting TiDE
axs[1].plot(pred_input['date'], pred_input['median'], label='Input', linewidth=3)
axs[1].plot(test_series['date'][-pred_steps:], test_series['median'][-pred_steps:], label='Ground Truth', linewidth=3)
axs[1].plot(pred_series_tide['date'], pred_series_tide['median'], label='TiDE', linewidth=3)
axs[1].set_title('TiDE', fontweight='bold')
axs[1].set_xlabel('Timestamp (H)')
axs[1].set_ylabel('Price (USD)')
axs[1].legend()
axs[1].grid(False, axis='x')
axs[1].set_xlim(start_date, end_date)

# Plotting TCN
axs[2].plot(pred_input['date'], pred_input['median'], label='Input', linewidth=3)
axs[2].plot(test_series['date'][-pred_steps:], test_series['median'][-pred_steps:], label='Ground Truth', linewidth=3)
axs[2].plot(pred_series_tcn['date'], pred_series_tcn['median'], label='TCN', linewidth=3)
axs[2].set_title('TCN', fontweight='bold')
axs[2].set_xlabel('Timestamp (H)')
axs[2].set_ylabel('Price (USD)')
axs[2].legend()
axs[2].grid(False, axis='x')
axs[2].set_xlim(start_date, end_date)


# Tight layout
plt.tight_layout()

plt.savefig('crypto_prediction.png')
```
